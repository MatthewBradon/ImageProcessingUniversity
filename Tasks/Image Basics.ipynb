```python
%matplotlib inline

import cv2 # The OpenCV library; install using `pip install opencv-contrib-python`
import numpy as np # Helpful when working with arrays; install using `pip install numpy`
from matplotlib import pyplot as plt # Good for graphing; install using `pip install matplotlib`
from matplotlib import image as image
import easygui # An easy-to-use file-picker; pip install easygui (mac not supported)
```

```python
I = cv2.imread("../Images/Zardoz.jpg")

original_I = I.copy()

# I = cv2.cvtColor(I, cv2.COLOR_BGR2RGB)

# Conver to YUV
I = cv2.cvtColor(I, cv2.COLOR_BGR2YUV)


# Split the image into Y, U, V channels
Y, U, V = cv2.split(I)

# Convert BGR to Grayscale
original_I = cv2.cvtColor(original_I, cv2.COLOR_BGR2GRAY)

# Display I and original_I
plt.figure(figsize=(10, 10))
plt.subplot(121)
plt.imshow(original_I, cmap='gray')
plt.title('Grayscale Image')
plt.subplot(122)
plt.imshow(I)
plt.title('YUV Image')

# Display Y, U, V channels
plt.figure(figsize=(10, 10))
plt.subplot(121)
plt.imshow(Y , cmap='gray')
plt.title('Y Channel')
plt.subplot(122)
plt.imshow(original_I, cmap='gray')
plt.title('Grayscale Image')

```

```python
def draw(event, x, y, flags, param):
    if event == cv2.EVENT_LBUTTONDOWN:
        cv2.rectangle(I, (x-50, y-50), (x+50, y+50), (0, 255, 0), 2)

        cv2.imshow('image', I)


I = cv2.imread("../Images/Zardoz.jpg")

# Show the image to draw
cv2.imshow('image', I)
cv2.setMouseCallback('image', draw)



# Wait for Q key to close the window
while True:
    if cv2.waitKey(1) & 0xFF == ord('q'):
        cv2.destroyAllWindows()
        break

cv2.waitKey(1) # Wait for a key press
```
